```python
import math 
import numpy as np 
from scipy import signal 
import matplotlib.pyplot as plt

# b - изменяемый параметр
c = 1
```

```python
p_num = 101 
# точки для проверки
vs = np.logspace(0, 2, p_num)
vs_ = - 1 * vs # значения b < 0
```

```python
# в системе три полюса, поэтому для каждого b храним значения в матрице размера p_num x 3
Ps = np.empty((p_num, 3), dtype = 'complex_')
Ps_ = np.empty((p_num, 3), dtype = 'complex_') # для хранения полюсов при b < 0
```

```python
for i in range(p_num): 
    # рассчитываем нули, полюсы и усиление для проверяемой системы 
    z,p,k = signal.tf2zpk([vs[i], 1], [vs[i]*c, vs[i], c, 1]) 
    # сохраняем полюсы 
    Ps[i, 0] = p[0] # значение этого полюса зависит от незменяющегося параметра с 
    Ps[i, 1] = p[1]
    Ps[i, 2] = p[2]  
    z_,p_,k_ = signal.tf2zpk([vs_[i], 1], [vs_[i]*c, vs_[i], c, 1])  
    Ps_[i, 0] = p_[0] 
    Ps_[i, 1] = p_[1]
    Ps_[i, 2] = p_[2]  
# b = 0
z0, p0, k0 = signal.tf2zpk([1], [c, 1])   
```

```python
# изобразим результаты на графиках ( b > 0)

fg_0,ax_0 = plt.subplots() 
ax_0.plot(np.real(Ps[:, 0]), np.imag(Ps[:,0]), '.-') 

fg_1,ax_1 = plt.subplots() 
ax_1.plot(np.real(Ps[:, 1]), np.imag(Ps[:,1]), '.-') 

fg_2,ax_2 = plt.subplots() 
ax_2.plot(np.real(Ps[:, 2]), np.imag(Ps[:,2]), '.-') 



plt.show()
```

```python
# b < 0
fg_0,ax_0 = plt.subplots() 
ax_0.plot(np.real(Ps_[:, 0]), np.imag(Ps_[:,0]), '.-') 

fg_1,ax_1 = plt.subplots() 
ax_1.plot(np.real(Ps_[:, 1]), np.imag(Ps_[:,1]), '.-') 

fg_2,ax_2 = plt.subplots() 
ax_2.plot(np.real(Ps_[:, 2]), np.imag(Ps_[:,2]), '.-') 



plt.show()
```

```python
for i in range(0, p_num, 10): 
    print('For b=={0:.2f} P_0==({1:.2f}, {2:.2f})'.format(vs[i], Ps[i, 0].real, Ps[i, 0].imag)) 
    print('For b=={0:.2f} P_1==({1:.2f}, {2:.2f})'.format(vs[i], Ps[i, 1].real, Ps[i, 1].imag))
    print('For b=={0:.2f} P_2==({1:.2f}, {2:.2f})'.format(vs[i], Ps[i, 2].real, Ps[i, 2].imag))
 
# Видно, что при всех b > 0 действительные части P_1 и P_2 равны нулю => система неустойчива
```

```python
for i in range(0,p_num,10): 
    print('For b=={0:.2f} P_0==({1:.2f}, {2:.2f})'.format(vs_[i],Ps_[i, 0].real, Ps_[i, 0].imag)) 
    print('For b=={0:.2f} P_1==({1:.2f}, {2:.2f})'.format(vs_[i],Ps_[i, 1].real, Ps_[i, 1].imag))
    print('For b=={0:.2f} P_2==({1:.2f}, {2:.2f})'.format(vs_[i],Ps_[i, 2].real, Ps_[i, 2].imag))
 # Для всех b < 0 у полюсов присутствует неотрицательная действительная часть  => система неустойчива
```

```python
# b = 0 
print('For b==0 P_0== (', p0[0].real,',', p0[0].imag, ')') 
```

```python
# реакция системы на конечный сигнал при параметре, делающем систему нейстойчивой 
b = 10

freq = 0.1

test_RC = signal.lti([b, 1], [b*c, b, c, 1])
T = np.linspace(0, 10.0/freq, 1001)
S = [math.sin(t*2*math.pi*freq) for t in T] 
Tout, yout, xout = signal.lsim(test_RC, S, T) 

plt.figure() 
plt.plot(Tout, yout) 
plt.show()
```

```python
# АЧХ 
W, mag, phase = signal.bode(test_RC)
fs = [0.5 * w / math.pi for w in W]

plt.figure()
plt.semilogx(fs, mag)
plt.show()
```

```python
# ФЧХ 
plt.semilogx(fs, phase)
plt.show()
```

```python
# Замкнутая система W(s) = (1 + bs)/(2 + (b + c)s + bs^2 + bcs^3)

Ps_closed = np.empty((p_num, 3), dtype = 'complex_')
Ps_closed_ = np.empty((p_num, 3), dtype = 'complex_') 

for i in range(p_num): 
    # рассчитываем нули, полюсы и усиление для проверяемой системы 
    z,p,k = signal.tf2zpk([vs[i], 1], [vs[i]*c, vs[i], vs[i] + c, 2]) 
    # сохраняем полюсы 
    Ps_closed[i, 0] = p[0] # значение этого полюса зависит от незменяющегося параметра с 
    Ps_closed[i, 1] = p[1]
    Ps_closed[i, 2] = p[2]  
    z_,p_,k_ = signal.tf2zpk([vs_[i], 1], [vs_[i]*c, vs_[i], c, 1]) 
    Ps_closed_[i, 0] = p_[0] 
    Ps_closed_[i, 1] = p_[1]
    Ps_closed_[i, 2] = p_[2]  
# b = 0
z0, p0, k0 = signal.tf2zpk([1], [c, 1])   
```

```python
# изобразим результаты на графике ( b > 0)

fg_0,ax_0 = plt.subplots() 
ax_0.plot(np.real(Ps_closed[:, 0]), np.imag(Ps_closed[:,0]), '.-') 

fg_1,ax_1 = plt.subplots() 
ax_1.plot(np.real(Ps_closed[:, 1]), np.imag(Ps_closed[:,1]), '.-') 

fg_2,ax_2 = plt.subplots() 
ax_2.plot(np.real(Ps_closed[:, 2]), np.imag(Ps_closed[:,2]), '.-') 



plt.show()
```

```python
# изобразим результаты на графике ( b > 0)

fg_0,ax_0 = plt.subplots() 
ax_0.plot(np.real(Ps_closed_[:, 0]), np.imag(Ps_closed_[:,0]), '.-') 

fg_1,ax_1 = plt.subplots() 
ax_1.plot(np.real(Ps_closed_[:, 1]), np.imag(Ps_closed_[:,1]), '.-') 

fg_2,ax_2 = plt.subplots() 
ax_2.plot(np.real(Ps_closed_[:, 2]), np.imag(Ps_closed_[:,2]), '.-') 



plt.show()
```

```python
for i in range(0, p_num, 10): 
    print('For b=={0:.2f} P_0==({1:.2f}, {2:.2f})'.format(vs[i], Ps_closed[i, 0].real, Ps_closed[i, 0].imag)) 
    print('For b=={0:.2f} P_1==({1:.2f}, {2:.2f})'.format(vs[i], Ps_closed[i, 1].real, Ps_closed[i, 1].imag))
    print('For b=={0:.2f} P_2==({1:.2f}, {2:.2f})'.format(vs[i], Ps_closed[i, 2].real, Ps_closed[i, 2].imag))
 
# Устойчива при b > 1, на границе устойчиовсти при b = 1 = c, что соответствует результатам, полученным критериями Рауса и Гурвица:
# b > 0, b > c,  (c > 0)
```
